```python
import pandas as pd
import numpy as np

credits=pd.read_csv("tmdb_5000_credits.csv")
movies=pd.read_csv("tmdb_5000_movies.csv")
```

```python
#combine the two dataframes with respect to id since they are same in both columns
credits_rename=credits.rename(index=str,columns={"movie_id":"id"})
movie_merge=movies.merge(credits_rename,on='id')
movie_merge.head()
```

```python
#remove unnecessary columns
movies_clean_data=movie_merge.drop(columns=['homepage','title_x','title_y','status','production_countries'])
movies_clean_data.head()
```

```python
movies_clean_data.info()
```

Weighted average for each movie

w=(Rv+Cm)/(v+m)

W=weighted rating
R=average for movie as a number from 0 - 10 (mean)=(Ratings)
v=no of votes for movie
m=min votes req to be listed in top 250 (currently 3000)
C=mean vote across the whole report (currently 6.9)


```python
#calculate all values  from the above formula
v=movies_clean_data['vote_count']
R=movies_clean_data['vote_average']
C=movies_clean_data['vote_average'].mean()
m=movies_clean_data['vote_count'].quantile(0.70) # Movies with more than 70 percentile
```

```python
movies_clean_data['weighted_average']=((R*v)+(C*m))/(v+m)
movies_clean_data.head()
```

```python
movies_ranking=movies_clean_data.sort_values('weighted_average',ascending=False)
movies_ranking[['original_title','vote_count','vote_average','weighted_average','popularity']].head(10)
```

Content Based Recommendation System - Now lets make a recommendations based on the movie’s plot summaries given in the overview column. So if our user gives us a movie title, our goal is to recommend movies that share similar plot summaries.

```python
movies_clean_data.head(10)['overview']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfv = TfidfVectorizer(min_df=3,  max_features=None, 
            strip_accents='unicode', analyzer='word',token_pattern=r'\w{1,}',
            ngram_range=(1, 3),
            stop_words = 'english')

# Filling NaNs with empty string
movies_clean_data['overview'] = movies_clean_data['overview'].fillna('')
```

```python

# Fitting the TF-IDF on the 'overview' text
tfv_matrix = tfv.fit_transform(movies_clean_data['overview'])
```

```python
tfv_matrix.shape
```

```python
from sklearn.metrics.pairwise import sigmoid_kernel

# Compute the sigmoid kernel
sig = sigmoid_kernel(tfv_matrix, tfv_matrix)
```

```python
# Reverse mapping of indices and movie titles
indices = pd.Series(movies_clean_data.index, index=movies_clean_data['original_title']).drop_duplicates()
```

```python
indices
```

```python
def give_rec(title, sig=sig):
    # Get the index corresponding to original_title
    idx = indices[title]

    # Get the pairwsie similarity scores 
    sig_scores = list(enumerate(sig[idx]))

    # Sort the movies 
    sig_scores = sorted(sig_scores, key=lambda x: x[1], reverse=True)

    # Scores of the 10 most similar movies
    sig_scores = sig_scores[1:11]

    # Movie indices
    movie_indices = [i[0] for i in sig_scores]

    # Top 10 most similar movies
    return movies_clean_data['original_title'].iloc[movie_indices]
```

```python
# Testing our content-based recommendation system with the seminal film Spy Kids
give_rec('Spy Kids')
```
